B. CUSTOMER ANALYSIS


-Who are the top 10 customers by total revenue?


```python
Top_ten = sales.groupby('CustomerID')['Revenue'].sum().sort_values(ascending=False).head(10)
Top_ten_df = Top_ten.reset_index()
Top_ten_df.columns = ['CustomerID', 'TotalRevenue']
print(Top_ten_df)

```

-Which customers buy most frequently (number of orders)?

```python
orders_per_customer = sales.groupby('CustomerID')['InvoiceNo'].nunique()
top_customers = orders_per_customer.sort_values(ascending=False).head(10)
top_customers_df = top_customers.reset_index()
top_customers_df.columns = ['CustomerID', 'NumberOfOrders']
print(top_customers_df)
```

-How many customers made only one purchase?

```python
total_one_purchase = (sales.groupby('CustomerID')['InvoiceNo'].nunique() == 1).sum()
print(total_one_purchase)
```

-What is the distribution of customer revenue (e.g., low, medium, high)?


```python
customer_revenue = sales.groupby('CustomerID')['Revenue'].sum()
customer_segments = pd.qcut(customer_revenue, q=3, labels=['Low', 'Medium', 'High'])
segment_distribution = customer_segments.value_counts()
print(segment_distribution)
```

Customers are grouped based on the total revenue they generate. 
They are classified into low, medium, and high revenue customers to better understand spending behavior.
